# Selection

```python
meta_cols = ["lr", "scheduler", "reduce_on_plateau_patience", "early_stop_patience"]
cols_to_avg = ["S01_mean_acc","S02_mean_acc","S03_mean_acc","S04_mean_acc","mean_acc_over_subjects"]
subject_cols = ["S01_mean_acc", "S02_mean_acc", "S03_mean_acc", "S04_mean_acc"]
```

```python
files_vocalized = [r"/scratch2/gspacone/sensors_2026_speechnet_rest_seed2/speech_net_base_vocalized_lr_ablations.csv", 
              r"/scratch2/gspacone/sensors_2026_speechnet_rest_seed2/speech_net_base_vocalized_lr_ablations.csv"]
files_silent = [r"/scratch2/gspacone/sensors_2026_speechnet_rest_seed2/speech_net_base_silent_lr_ablations.csv", 
              r"/scratch2/gspacone/sensors_2026_speechnet_rest_seed2/speech_net_base_silent_lr_ablations.csv"]

dfs_vocalized= []
dfs_silent = []
for file in files_vocalized:
    dfs_vocalized.append(pd.read_csv(file))
for file in files_silent:
    dfs_silent.append(pd.read_csv(file))
```

```python
# merge across condition (silent or vocalized), average seeds
dfs_vocalized_final = pd.concat(dfs_vocalized, ignore_index=True)
df_avg_voc = (dfs_vocalized_final.groupby("variant", as_index=False)[cols_to_avg].mean())
df_meta_voc = (dfs_vocalized_final.groupby("variant", as_index=False)[meta_cols].first())
df_final_vocalized = df_avg_voc.merge(df_meta_voc, on="variant")
df_subj_std = df_final_vocalized[subject_cols].std(axis=1, ddof=0)
df_final_vocalized["std_across_subjects"] = df_subj_std
df_final_vocalized.sort_values(by='mean_acc_over_subjects',ascending=False)
```

```python
# merge across condition (silent or vocalized), average seeds
dfs_silent_final = pd.concat(dfs_silent, ignore_index=True)
df_avg_silent = (dfs_silent_final.groupby("variant", as_index=False)[cols_to_avg].mean())
df_meta_silent = (dfs_silent_final.groupby("variant", as_index=False)[meta_cols].first())
df_final_silent = df_avg_silent.merge(df_meta_silent, on="variant")
df_subj_std = df_final_silent[subject_cols].std(axis=1, ddof=0)
df_final_silent["std_across_subjects"] = df_subj_std
df_final_silent.sort_values(by='mean_acc_over_subjects',ascending=False)
```

```python
# do the same for silent and vocalized

df_both_cond = pd.concat((df_final_vocalized, df_final_silent))
# group by conditions
df_avg_conds = ((df_both_cond.groupby("variant", as_index=False))[cols_to_avg].mean())
df_meta_both = ((df_both_cond.groupby("variant", as_index=False))[meta_cols].first())
df_final_both = df_avg_conds.merge(df_meta_both, on="variant")

df_std_both = df_final_both[subject_cols].std(axis=1, ddof=0)
df_final_both["std_across_subjects"] = df_std_both
df_final_both.sort_values(by='mean_acc_over_subjects', ascending=False)
```
